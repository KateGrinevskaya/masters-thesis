```python
import os
import pandas as pd
```

```python
current_dir = os.getcwd()
```

```python
def cases_in_terms(string):
    result = ''
    pieces = string.split(' ')
    words = []
    for piece in pieces:
        if piece:
            dashed = piece.split('-')
            if len(dashed) == 1:
                dashed_res = dashed[0]
                dashed_res = dashed_res[0].lower() + dashed_res[1::]
            else:
                dashed_res = []
                for d in dashed:
                    d_cased = d[0].lower() + d[1::]
                    dashed_res.append(d_cased)
                dashed_res = '-'.join(dashed_res)
            words.append(dashed_res)
    result = ' '.join(words)
    return result
```

```python
def dashes_in_defs(definition):
    parts = definition.split('—')
    if len(parts) == 1:
        return definition
    else:
        needed_parts = parts[1::]
        return '—'.join(needed_parts)
```

```python
def write_to_lists(list_to_write_terms, list_to_write_defs, file):
    with open (file, 'r', encoding='utf-8') as f:
        text = f.read()
        pairs = text.split('term: ')
        for pair in pairs:
            if pair:
                term, definition = pair.split(', definition: ')[0].strip(), pair.split(', definition: ')[-1].strip()
                term = cases_in_terms(term).replace('\n', ' ')
                definition = definition[0].upper() + definition[1::].replace('\n', ' ')
                definition = dashes_in_defs(definition)
                if term and definition:
                    list_to_write_terms.append(term)
                    list_to_write_defs.append(definition)
    return list_to_write_terms, list_to_write_defs
```

```python
eng_terms = []
eng_defs = []
ru_terms = []
ru_defs = []
for root, dirs, files in os.walk(current_dir):
    for file in files:
        if file.endswith('.txt'):
            if file.startswith('eng_'):
                write_to_lists(eng_terms, eng_defs, os.path.join(root, file))
            else:
                write_to_lists(ru_terms, ru_defs, os.path.join(root, file))
eng_df = pd.DataFrame({'term': eng_terms, 'definition': eng_defs})
ru_df = pd.DataFrame({'term': ru_terms, 'definition': ru_defs})
```
