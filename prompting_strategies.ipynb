# Load model

```python
import torch
from transformers import AutoModelForCausalLM, AutoTokenizer, pipeline
```

```python
torch.random.manual_seed(RANDOM_SEED)
```

```python
model_path = "microsoft/Phi-4-mini-instruct"

model_phi = AutoModelForCausalLM.from_pretrained(
    model_path,
    device_map="auto",
    torch_dtype="auto",
    trust_remote_code=True,
)
tokenizer_phi = AutoTokenizer.from_pretrained(model_path)
```

```python
device = torch.device("cuda:0" if torch.cuda.is_available() else "cpu")
```

```python
model_phi = model_phi.to(device)
```

# Check model

```python
def translate(model, tokenizer, input_text):
    messages = [
      {"role": "system", "content": "You are a professional translator."},
      {"role": "user", "content": input_text},
    ]

    pipe = pipeline(
        "text-generation",
        model=model,
        tokenizer=tokenizer,
    )

    generation_args = {
        "max_new_tokens": 500,
        "return_full_text": False,
        "temperature": 0.0,
        "do_sample": False
    }

    output = pipe(messages, **generation_args)
    return output[0]['generated_text']
```

```python
PROMPT_START = """
English: Interjections are a fairly common part of the vocabulary of the modern Bashkir language.
Russian: Междометия представляют собой довольно распространенный пласт лексики современного башкирского языка.

English: The need to describe cases in the Khanty language is due to the insufficient research of this issue in modern Finno-Ugric studies.
Russian: Необходимость описания падежей в хантыйском языке обусловлена недостаточным исследованием данного вопроса в современном финно-угроведении.

English: The work is devoted to the analysis of dialect words of phonemic type, recorded in the territory of the Penza region; words with lexicalized phonetic features are considered.
Russian: Работа посвящена анализу диалектных слов фонематического типа, зафиксированных на территории Пензенской области; рассматриваются слова с лексикализованными фонетическими особенностями.

"""
```

```python
text = """
English: The focus of this article is a comparative analysis of idioms that include the lexeme weather in Russian and weather in English, with the aim of studying the national specifics of the perception of the phenomenon weather.
Russian:
"""
```

```python
translate(model_phi, tokenizer_phi, PROMPT_START+text)
```

```python
def postprocess_output(text):
    if 'Russian: ' in text:
        return text.split('Russian: ')[-1]
    return text
```

# 3-shot prompting

```python
test_src_texts = test['en']
prompts = []
for sent in test_src_texts:
    df_elements = train.sample(n=3)
    PROMPT_START = f"""
    English: {df_elements.iloc[0]['en']}
    Russian: {df_elements.iloc[0]['ru']}

    English: {df_elements.iloc[1]['en']}
    Russian: {df_elements.iloc[1]['ru']}

    English: {df_elements.iloc[2]['en']}
    Russian: {df_elements.iloc[2]['ru']}

    """
    text_to_translate = f"""
    English: {sent}
    Russian:
    """
    prompt_text = PROMPT_START + text_to_translate
    prompts.append(prompt_text)
test_tgt_texts = test['ru']
predictions_phi = []
```

```python
batch_size = 16

for i in tqdm(range(0, len(prompts), batch_size)):
    batch_src = prompts[i:i + batch_size]
    preds = translate(model_phi, tokenizer_phi, batch_src)
    predictions_phi.extend(postprocess_output(preds))
```

```python
with open('predictions_phi_prompt_strat.txt', 'w') as f:
    for line in predictions_phi:
        f.write(f"{line}\n")
```

```python
result_prompt_strat = {
    'bleu': blue.compute(predictions=predictions_phi, references=test_tgt_texts.tolist()),
    'comet': comet.compute(predictions=predictions_phi, references=test_tgt_texts.tolist(), sources=test_src_texts.tolist())['mean_score'],
    'chrf': chrf.compute(predictions=predictions_phi, references=test_tgt_texts.tolist())['score'],
    'rouge': rouge.compute(predictions=predictions_phi, references=test_tgt_texts.tolist())
}

print(result_prompt_strat)
```

# Providing linguistic glossary

```python
import re
```

```python
russian_pairs = pd.read_csv('russian_pairs.csv').drop(columns=['Unnamed: 0'])
english_pairs = pd.read_csv('english_pairs.csv').drop(columns=['Unnamed: 0'])
```

```python
en_terms = english_pairs['term']
```

```python
term_defs_prompts_for_sents = []
en_terms = english_pairs['term']
for sent in test_src_texts:
    term_defs = []
    sent_lower = sent.lower()
    terms_in_sent = set([term for term in en_terms if term in sent_lower])
    for term in terms_in_sent:
        found_defin = english_pairs.loc[english_pairs['term'] == term]['definition']
        if len(found_defin) == 1:
            found_defin = found_defin.item()
        else:
            found_defin = found_defin.to_frame()
            found_defin = found_defin.iloc[0].item()
        row_with_terms_defs = 'term: ' + term + '; definition: ' + found_defin
        term_defs.append(row_with_terms_defs)
    if term_defs:
        term_def_pairs = '; '.join(term_defs)
        sent_with_term_def = 'Translate ONLY the following sentence to Russian: ' + sent + term_def_pairs
    else:
        sent_with_term_def = PROMPT_START + sent
    term_defs_prompts_for_sents.append(sent_with_term_def)
```

```python
term_defs_prompts_for_sents[151]
```

```python
batch_size = 64
predictions_phi_dicts = []

for i in tqdm(range(0, len(term_defs_prompts_for_sents), batch_size)):
    batch_src = prompts[i:i + batch_size]
    preds = translate(model_phi, tokenizer_phi, batch_src)
    predictions_phi_dicts.extend(postprocess_output(preds))
```

```python
result_dicts = {
    'bleu': blue.compute(predictions=predictions_phi_dicts, references=test_tgt_texts),
    'comet': comet.compute(predictions=predictions_phi_dicts, references=test_tgt_texts, sources=test_src_texts)['mean_score'],
    'chrf': chrf.compute(predictions=predictions_phi_dicts, references=test_tgt_texts)['score'],
    'rouge': rouge.compute(predictions=predictions_phi_dicts, references=test_tgt_texts)
}

print(result_dicts)
```
